# N-Body Problem Solver

#### This solver calculates an intercept trajectory for DART-styled satellite missions. While this solver simply implements the fastest path to intercept with a set burn time, it can be adapted to accomplish other missions such as landing on the asteroid, colliding at the greatest speed, etc. Inputting the parameters required, the user will obtain design parameters for a transfer burn: necessary fuel mass and optimal thrust vector.

Please input bodies and their data below, following the format provided for Earth, the Moon, and the Sun. The solver assumes that mass flow rate of fuel is constant, and that propulsion velocity is constant, neglecting actions such as orienting before setting off on the main transfer burn. It also assumes that no thrust vectoring will occur besides stabilizing pulses, which are considered negligible, and that once the transfer burn is started, it exerts a constant force on the satellite.

This solver considers an Earth-centered frame of reference, with the origin's cartesian axes being oriented as such:
1. X-axis: vector from Earth's center to the longitude/latitude of 0°, 0°.
2. Y-axis: vector from Earth's center to the North Pole (longitude 90°, latitude doesn't matter).
3. Z-axis: vector from Earth's center to the longitude/latitude of 0°, 90°E
